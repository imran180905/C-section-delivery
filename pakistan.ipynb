```python
def do_the_job(x):
    
    if(x<=2.0):
        return 0
        
    elif (x==3 or x==4):
        return 1       
        
        
    elif(x>4):
        return 2    
    # x==0 means mother's age is between 15-24 year 
     # x==1 means mother's age is between 25-34 year
      # x==2 means mother's age is between 35-49 year                   
```

```python
df['Age of the Mother']= df.V013.apply(do_the_job)
```

```python
def do_the_job1(x):
    ret=2.0   ## If x=2, Overweight
    if(x< 1850.0):
        ret=0.0   ## If x=0, Under weight
        
    elif (1850.0<=x<=2500.0):
        ret=1.0  ## If x=1, Normal weight
        
        
    return ret    
```

```python
df['Mother’s BMI']= df.V445.apply(do_the_job1)
```

```python
def do_the_job5(x):
             
    if(x==1):
        
        return 0       
        
    elif (x==2): 
        
        return 1       
        
        
    elif(x==4):
        
        return 2
    
    elif(x==3 or x==5 or x==6):
        return 3
 # X=0 means mother herself takes the decision
# X=1 means both husband & wife take decision
# X=2 means only husband alone take the decision
# X=3 means someone else take decision
    
```

```python
def do_the_job6(x):
             
    if(x==0):
        
        return 0       
        
    elif (x==4 or x==5): 
        
        return 1       
        
        
    elif(x==1):
        
        return 2
    
    elif(x==3 or x==2):
        
        return 3
    elif(6<=x<=9):
        return 4
    
# X=0 means husband has no work    
# X=1 means husband's work is agricultural
# X=2 means husband is professional
# X=3 means husband is business man
# X=4 means other occupation
```
